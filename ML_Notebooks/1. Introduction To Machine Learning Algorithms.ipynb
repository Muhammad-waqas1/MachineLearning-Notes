# 1. Introduction To Machine Learning Algorithms

AI vs ML vs DL vs DS
---------------------

1. AI : It can do its tasks without any Human interventions
- Netflix
- Self-Driving
- Amazon.in

2. ML : Is Subset of AI that provides stats tools to analyze, visualize, forecasting and to do predictions

3. DL : Is Subset of ML that mimic the human brain using multilayer neural network

4. DS : Is Part of Everything. Responsive creating an AI application using any method possible
